```python
import rasterio
from rasterio.plot import show
import matplotlib.pyplot as plt
import numpy as np
```

```python
base_filename = "06_10_2024_F120_CottonB67_M3E_15m_RGB_MS_transparent_mosaic_"

file_rgb = base_filename+"group1.tif"
img_rgb = rasterio.open(file_rgb)
show(img_rgb)

file_nir = base_filename+"nir.tif"
img_nir = rasterio.open(file_nir)
show(img_nir)

file_re = base_filename+"red edge.tif"
img_re = rasterio.open(file_re)
show(img_re)
```

```python
num_bands_rgb = img_rgb.count
print(num_bands_rgb)

num_bands_nir = img_nir.count
print(num_bands_nir)

num_bands_re = img_re.count
print(num_bands_re)
```

```python
# Read the bands
band1_rgb = img_rgb.read(1)
band2_rgb = img_rgb.read(2)
band3_rgb = img_rgb.read(3)
band4_rgb = img_rgb.read(4)

# Display the band using matplotlib
plt.imshow(band1_rgb, cmap='gray')  # You can choose a different colormap if needed
plt.title("First Band - RGB")
plt.colorbar(label="Pixel Value")
plt.show()

# Display the band using matplotlib
plt.imshow(band2_rgb, cmap='gray')  # You can choose a different colormap if needed
plt.title("Second Band - RGB")
plt.colorbar(label="Pixel Value")
plt.show()

# Display the band using matplotlib
plt.imshow(band3_rgb, cmap='gray')  # You can choose a different colormap if needed
plt.title("Third Band - RGB")
plt.colorbar(label="Pixel Value")
plt.show()

# Display the band using matplotlib
plt.imshow(band4_rgb, cmap='gray')  # You can choose a different colormap if needed
plt.title("Fourth Band - RGB")
plt.colorbar(label="Pixel Value")
plt.show()

# Read the bands
band1_nir = img_nir.read(1)
band2_nir = img_nir.read(2)

# Display the band using matplotlib
plt.imshow(band1_nir, cmap='gray')  # You can choose a different colormap if needed
plt.title("First Band - NIR")
plt.colorbar(label="Pixel Value")
plt.show()

# Display the band using matplotlib
plt.imshow(band2_nir, cmap='gray')  # You can choose a different colormap if needed
plt.title("Second Band - NIR")
plt.colorbar(label="Pixel Value")
plt.show()

# Read the bands
band1_re = img_re.read(1)
band2_re = img_re.read(2)

# Display the band using matplotlib
plt.imshow(band1_re, cmap='gray')  # You can choose a different colormap if needed
plt.title("First Band - RE")
plt.colorbar(label="Pixel Value")
plt.show()


# Display the band using matplotlib
plt.imshow(band2_re, cmap='gray')  # You can choose a different colormap if needed
plt.title("Second Band - RE")
plt.colorbar(label="Pixel Value")
plt.show()
```

```python
# Use the RGB image metadata as a reference for the outputs
meta = img_rgb.meta

# We want a total of 5 bands
meta.update(count=5)

# Stack the bands; Reorder these as necessary to fit the desired style
stacked_bands = [band1_rgb, band2_rgb, band3_rgb, band1_nir, band1_re]

# Create a 3D array with stacked bands
multiband_array = np.stack(stacked_bands, axis=0)

# Write the multiband array to a new georeferenced TIFF file
with rasterio.open('multiband_output.tif', 'w', **meta) as dest:
    dest.write(multiband_array)
    
print("Done")
```
